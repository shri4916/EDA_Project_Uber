**UBER DATA ANALYTICS EDA PROJECT**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# To read a CSV file
df = pd.read_csv('ncr_ride_bookings.csv')

# View the first 10 rows
df.head(10)
```

```python
# Summary of numeric columns
df.describe()
```

```python
# information about the dataset's structure, column data types, and memory usage.
df.info()
```

```python
# Replace the null values with empty strings.
uber = df.fillna('')
uber
```

```python
# Pie chart to show number of transactions for each payment method

pay = df['Payment Method'].value_counts()
pay = pd.DataFrame(pay).reset_index()
print(pay)
pay.columns = ['Payment Method','no. of transactions']
fig = px.pie(pay,names='Payment Method',values='no. of transactions')
fig.show()
```

```python
# Number of customers per vehicle type

Vehicle_Type = df["Vehicle Type"].value_counts()
Vehicle_Type = pd.DataFrame(Vehicle_Type).reset_index()
Vehicle_Type.columns = ["Vehicle Type","Number of customers"]
print(Vehicle_Type)
fig = px.bar(x='Vehicle Type',y='Number of customers',data_frame=Vehicle_Type )
fig.show()
```

```python
veh=df.groupby('Vehicle Type')['Customer Rating'].mean()
veh=pd.DataFrame(veh).reset_index()
veh.columns=['Vehicle Type','Customer Rating']
print(veh)
fig=px.line(x='Vehicle Type',y='Customer Rating',data_frame=veh)
fig.show()
```

```python
# monthly customers
df['Date'] = pd.to_datetime(df['Date'])

monthly = df['Date'].dt.to_period("M").value_counts().sort_index().reset_index(name='customers')
monthly.columns=['Month','Customers']
print(monthly)
monthly['Month']=monthly['Month'].astype(str)
fig=px.bar(x='Month',y='Customers',data_frame=monthly, title="Monthly customers")
fig.show()

```

```python
# calculate total booking value, success booking value, avg distance travelled, total distance travelled per vehicle type.

veh_stats = df.groupby('Vehicle Type').agg(
    Total_Booking_Value=('Booking Value', 'sum'),
    Avg_Distance_Travelled=('Ride Distance', 'mean'),
    Total_Distance_Travelled=('Ride Distance', 'sum')
).reset_index()

# Calculate success booking value separately
success_bookings = df[df['Booking Status'] == 'Completed'].groupby('Vehicle Type')['Booking Value'].sum()
veh_stats['Success_Booking_Value'] = veh_stats['Vehicle Type'].map(success_bookings).fillna(0)

# Reorder columns
veh_stats = veh_stats[['Vehicle Type', 'Total_Booking_Value', 'Success_Booking_Value',
                        'Avg_Distance_Travelled', 'Total_Distance_Travelled']]

veh_stats
```

```python
# Calculate average customer and driver ratings per vehicle type
rating_stats = df.groupby('Vehicle Type').agg(
    Avg_Customer_Rating=('Customer Rating', 'mean'),
    Avg_Driver_Rating=('Driver Ratings', 'mean')
).reset_index()

rating_stats
```

```python
# Filter and count cancelled rides by customers
cancelled_rides = df[df['Booking Status'] == 'Cancelled by Customer']
cancellation_counts = cancelled_rides['Reason for cancelling by Customer'].value_counts()
cancel_customer = pd.DataFrame(cancellation_counts).reset_index()
print(cancel_customer)
cancel_customer.columns = ['Reason for cancelling by Customer', 'Count']
fig = px.pie(cancel_customer, names='Reason for cancelling by Customer', values='Count')
fig.show()
```

```python
# Filter and count cancelled rides by Drivers

cancelled_rides = df[df['Booking Status'] == 'Cancelled by Driver']
cancellation_counts = cancelled_rides['Driver Cancellation Reason'].value_counts()
cancel_driver = pd.DataFrame(cancellation_counts).reset_index()
print(cancel_driver)
cancel_driver.columns = ['Driver Cancellation Reason', 'Count']
fig = px.pie(cancel_driver, names='Driver Cancellation Reason', values='Count')
fig.show()
```
